```python
data['Year']=data['Release_Date'].dt.year
data['Year'].dtype
```

```python
data.drop(['Overview','Original_Language','Poster_Url'],axis=1,inplace=True)
data.columns
```

```python
def data_category(data,col,labels):
  check = [
      data[col].describe()['min'] if 'min' in data[col].describe() else data[col].min(),  # Use .min() if 'min' key is not present
      data[col].describe()['25%'] if '25%' in data[col].describe() else data[col].quantile(0.25),  # Use .quantile() if '25%' key is not present
      data[col].describe()['50%'] if '50%' in data[col].describe() else data[col].quantile(0.50),  # Use .quantile() if '50%' key is not present
      data[col].describe()['75%'] if '75%' in data[col].describe() else data[col].quantile(0.75),  # Use .quantile() if '75%' key is not present
      data[col].describe()['max'] if 'max' in data[col].describe() else data[col].max()   # Use .max() if 'max' key is not present
  ]
  data[col]=pd.cut(data[col],check,labels=labels,duplicates='drop').astype(str)
  return data
```

```python
labels=['Not Popular','Below Average','Average','Popular']
```

```python
data_category(data,'Vote_Average',labels)
```
